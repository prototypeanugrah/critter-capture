```python
from dataclasses import dataclass
from typing import List


@dataclass
class ObservationRecord:
    uuid: str
    image_url: str
    label_index: int


@dataclass
class DatasetBundle:
    train: ObservationsDataset
    validation: ObservationsDataset
    test: ObservationsDataset
    label_names: Sequence[str]
    label_ids: Sequence[int]


def _extract_image_url(row: pd.Series, image_url_column) -> str | None:
    def _clean(value) -> str | None:
        if value is None or pd.isna(value):
            return None
        text = str(value).strip()
        if not text or text.lower() == "nan":
            return None
        return text

    primary = _clean(row.get(image_url_column))
    if primary:
        return primary

    return None


def _load_records(
    df,
    uuid_column,
    image_url_column,
    label_column,
    label_names_column,
) -> tuple[List[ObservationRecord], List[str], List[int]]:
    required_columns = {
        uuid_column,
        image_url_column,
        label_column,
        label_names_column,
    }
    missing = required_columns - set(df.columns)
    if missing:
        raise ValueError(f"Missing required columns in CSV: {missing}")

    essential_columns = {
        uuid_column,
        label_column,
        label_names_column,
    }
    df = df.dropna(subset=list(essential_columns))

    df[label_column] = df[label_column].astype(int)

    label_id_to_name = (
        df[[label_column, label_names_column]]
        .drop_duplicates(subset=[label_column])
        .set_index(label_column)[label_names_column]
        .to_dict()
    )
    label_ids_sorted = sorted(label_id_to_name.keys())
    label_names = [label_id_to_name[label_id] for label_id in label_ids_sorted]
    label_to_index = {label_id: idx for idx, label_id in enumerate(label_ids_sorted)}

    grouped = df.groupby(uuid_column, sort=False)
    records: List[ObservationRecord] = []
    for uuid, group in grouped:
        image_url = _extract_image_url(group.iloc[0], image_url_column)
        if not image_url:
            print("Skipping observation %s due to missing image URL.", uuid)
            continue
        label_ids_unique = sorted(
            {int(label) for label in group[label_column].tolist()}
        )
        if len(label_ids_unique) != 1:
            print(
                f"Observation {uuid} has {len(label_ids_unique)} labels; selecting the first for multi-class setup.",
            )
        label_id = label_ids_unique[0]
        label_index = label_to_index[label_id]
        records.append(
            ObservationRecord(
                uuid=uuid,
                image_url=image_url,
                label_index=label_index,
            )
        )

    print(f"Loaded {len(records)} unique observations with {len(label_names)} labels")
    return records, label_names, label_ids_sorted


records, label_names, label_ids = _load_records(
    inat_df,
    "uuid",
    "image_url",
    "taxon_id",
    "common_name",
)
```

```python
from typing import Tuple

from sklearn.model_selection import train_test_split


def stratified_train_val_test_split(
    labels: np.ndarray,
    validation_size: float,
    test_size: float,
    seed: int,
) -> Tuple[np.ndarray, np.ndarray, np.ndarray]:
    """Create stratified train/validation/test splits for multi-class data."""

    if validation_size + test_size >= 0.9:
        raise ValueError(
            "Combined validation and test size should leave reasonable train data."
        )

    indices = np.arange(len(labels))
    stratify_labels = labels if np.min(np.bincount(labels)) >= 2 else None
    train_idx, holdout_idx = train_test_split(
        indices,
        test_size=validation_size + test_size,
        random_state=seed,
        stratify=stratify_labels,
    )

    holdout_labels = labels[holdout_idx]
    stratify_holdout = (
        holdout_labels if np.min(np.bincount(holdout_labels)) >= 2 else None
    )
    val_idx, test_idx = train_test_split(
        holdout_idx,
        test_size=test_size / (validation_size + test_size),
        random_state=seed,
        stratify=stratify_holdout,
    )

    print(
        f"Created stratified splits: train={len(train_idx)}, val={len(val_idx)}, test={len(test_idx)}",
    )

    return train_idx, val_idx, test_idx


val_size, test_size = 0.15, 0.15
seed = 42

label_array = np.array(
    [record.label_index for record in records],
    dtype=np.int32,
)

train_idx, val_idx, test_idx = stratified_train_val_test_split(
    label_array,
    validation_size=val_size,
    test_size=test_size,
    seed=seed,
)
```

```python
from typing import Callable, List, Sequence

import torch
from PIL import Image
from torch.utils.data import Dataset as TorchDataset


class AnimalDataset(TorchDataset):
    def __init__(
        self,
        records: List[ObservationRecord],
        label_names: Sequence[str],
        transform: Callable,
    ):
        self.records = records
        self.label_names = label_names
        self._transform = transform

    def __len__(self):
        return len(self.records)

    def __getitem__(self, index: int):
        record = self.records[index]
        image = self.load_image(record.image_url)
        image_tensor = self._transform(image)
        target = torch.tensor(record.label_index, dtype=torch.long)
        return image_tensor, target

    def load_image(self, record: ObservationRecord) -> Image.Image:
        """Load an image from a URL."""
        cache_path = self._cache_dir / f"{record.uuid}.jpg"
        if not cache_path.exists():
            if not self._download_with_retries(
                record.image_url, cache_path, record.uuid
            ):
                return self._placeholder_image(cache_path)

        image = Image.open(cache_path)
        return image
```

```python
from torchvision import transforms as T


def transforms(image_size: int):
    normalize = T.Normalize(
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225],
    )

    train_transform = T.Compose(
        [
            T.Resize((image_size, image_size)),
            T.RandomHorizontalFlip(p=0.5),
            # T.RandomVerticalFlip(p=0.2),
            # T.ColorJitter(0.1, 0.1, 0.1, 0.05),
            T.ToTensor(),
            normalize,
        ]
    )
    eval_transform = T.Compose(
        [
            T.Resize((image_size, image_size)),
            T.ToTensor(),
            normalize,
        ]
    )

    return train_transform, eval_transform


train_transform, eval_transform = transforms(image_size=224)
```
